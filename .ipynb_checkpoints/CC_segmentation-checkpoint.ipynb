```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('customer_segmentation.csv')
```

```python
df
```

```python
df.describe()
```

```python
# check for missing data
pd.isna(df).sum()
```

```python
from pandas.api.types import is_numeric_dtype
numCols = [col for col in df.drop("customer_id", axis=1).columns if is_numeric_dtype(df[col])]
```

```python
fig, ax = plt.subplots(figsize=(12, 10))

df.drop('customer_id', axis=1).hist(ax=ax)

plt.tight_layout()
plt.show()
```

```python
catCols = [col for col in df.columns if not is_numeric_dtype(df[col])]
catCols
```

```python
df_mod = df.copy()
df_mod['gender'] = df_mod['gender'].apply(lambda x: 1 if x=='M' else 0)

ed_levels = {
    'Uneducated':0,
    'High School':1,
    'College':2,
    'Graduate':3,
    'Post-Graduate':4,
    'Doctorate':5}
df_mod['education_level'] = df_mod['education_level'].apply(lambda x: ed_levels[x])
```

```python
df_mod = pd.get_dummies(df_mod, columns=['marital_status'])
```

```python
df_mod
```

```python
df_scaled = df_mod.drop('customer_id',axis=1)

scaler = StandardScaler()
scaler.fit(df_scaled)
df_scaled = scaler.transform(df_scaled)
df_scaled = pd.DataFrame(df_scaled)

df_scaled
```

```python
inertias = []

for k in range(1,11):
    model = KMeans(n_clusters=k)
    clusters = model.fit_predict(df_scaled)
    inertias.append(model.inertia_)
```

```python
plt.plot(list(range(1,11)), inertias, marker = 'o')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouettes = []

for k in range(3,8):
    model = KMeans(n_clusters=k)

    X = df_scaled.values
    clusters = model.fit_predict(X)
    silScore = silhouette_score(X, clusters)
    silhouettes.append(silScore)
```

```python
model = KMeans(n_clusters=5)
clusters = model.fit_predict(df_scaled)
print(clusters)
```

```python
df['CLUSTER'] = clusters
for col in numCols:
    c_means = df.groupby('CLUSTER')[col].mean()
    plt.bar(c_means.index, c_means)
    plt.title(col)
    plt.xlabel('Cluster')
    plt.ylabel(f'Mean {col}')
    plt.show()
```

```python
focusPairs = [
    ['estimated_income', 'credit_limit'],
    ['estimated_income', 'total_trans_amount'],
    ['estimated_income', 'avg_utilization_ratio'],
    ['credit_limit', 'total_trans_amount'],
    ['credit_limit', 'avg_utilization_ratio']
]
```

```python
for pair in focusPairs:
    sns.scatterplot(data=df, x=pair[0], y=pair[1], hue='CLUSTER', palette='tab10')
    plt.show()
```

```python
for col in catCols:
    plot_df = pd.crosstab(
        index=df['CLUSTER'], columns=df[col],
      values=df[col], aggfunc='size', normalize='index'
    )
    
    plot_df.columns.name = None
    plot_df.index.name = None
    plot_df.reset_index(drop=False)
    
    plot_df.plot(kind='bar', stacked=True)
```
